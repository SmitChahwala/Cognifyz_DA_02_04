```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('Dataset.csv')
```

```python
df
```

```python
# Identify restaurant chains by grouping by 'Restaurant Name'
restaurant_chains = df['Restaurant Name'].value_counts()
restaurant_chains = restaurant_chains[restaurant_chains > 1].index.tolist()
```

```python
restaurant_chains
```

```python
# Filter the dataset to include only restaurant chains
df_chains = df[df['Restaurant Name'].isin(restaurant_chains)]
```

```python
df_chains
```

```python
# Analyze ratings and popularity (votes) of different restaurant chains
chain_analysis = df_chains.groupby('Restaurant Name').agg({
    'Aggregate rating': ['mean', 'count'],
    'Votes': 'sum'
}).reset_index()
```

```python
# Rename columns for clarity
chain_analysis.columns = ['Restaurant Name', 'Average Rating', 'Number of Outlets', 'Total Votes']
```

```python
# Sort by average rating and total votes
chain_analysis_sorted = chain_analysis.sort_values(by=['Average Rating', 'Total Votes'], ascending=False)
```
